```python
from selenium import webdriver
import schedule
import time
import sqlite3
from bs4 import BeautifulSoup

driver = webdriver.Chrome('./chromedriver.exe')
conn = sqlite3.connect('./saves/db.stock')
c = conn.cursor()
c.execute("CREATE TABLE IF NOT EXISTS article (id INTEGER PRIMARY KEY AUTOINCREMENT, title TEXT, content TEXT, time TEXT)")

uri = 'https://news.daum.net/breakingnews/?page='
# fnews = []

for page in range(1, 3):
    target = uri+str(page)
    driver.get(target)

    try:
        for n in range(1, 16):
#             time.sleep(2)
            driver.implicitly_wait(3)
            try: 
                driver.find_element_by_xpath('//*[@id="mArticle"]/div[2]/ul/li[%s]/div/strong/a' %n).click()
                html = driver.page_source
                soup = BeautifulSoup(html, 'html.parser')

                titles = soup.select('h3.tit_view')
                title = titles[0].text.strip()

                times = soup.select('span.num_date')
                time = times[0].text.strip()

                contents = soup.select('#harmonyContainer > section')
                content = contents[0].text.strip()

                driver.back()
                
#                 a = [title, content, time]
#                 fnews.append(a)

                c.execute("INSERT INTO article( title, content, time ) VALUES(?,?,?)",(title, content, time))

            except:
                pass

    except:
        pass  

conn.commit()
c.close()
# print(fnews)
```
